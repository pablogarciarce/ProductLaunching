```python
import numpy as np
from scipy.stats import dirichlet, gamma, uniform
from scipy.special import comb
import matplotlib.pyplot as plt
```

```python
def compute_expected_utility(t1, p1, lambda1, c11, c21, c31, n, T, ite=5000):
    pi = 0
    cost = 0

    for i in range(ite):
        def lambda23(t):
            a = np.random.normal(0.256, 0.05)
            if a < .01:
                a = .01  # nos aseguramos de que el valor sea positivo
            return a * t ** np.random.normal(0.837, 0.05)
        # Generating random variables for j=1
        e1 = np.random.poisson(lambda1(t1))
        eT = np.random.poisson(lambda1(T))
        q1 = eT - e1
        
        # Generating random variables for j=2,3
        ts = uniform.rvs(loc=0, scale=2000, size=2)
        ps = uniform.rvs(loc=3000, scale=15000, size=2)
        qs = np.random.poisson(lambda23(T)) - np.random.poisson(lambda23(ts))

        # Generating buyer random variables
        w = dirichlet.rvs([1, 1, 1])[0]
        rho = gamma.rvs(5, scale=1/5)

        # Computing utility
        u1 = 1 - np.exp(-rho * (-w[0] * t1 / T - w[1] * p1 / 15000 - w[2] * q1 / eT))
        u2 = 1 - np.exp(-rho * (-w[0] * ts[0] / T - w[1] * ps[0] / 15000 - w[2] * qs[0] / eT)) - u1
        u3 = 1 - np.exp(-rho * (-w[0] * ts[1] / T - w[1] * ps[1] / 15000 - w[2] * qs[1] / eT)) - u1

        # Computing probability of choice
        pi += 1 / (1 + np.exp(u2) + np.exp(u3))

        # Estimating the cost
        c1 = c11 * t1 + c21 * e1 + c31 * (eT - e1)
        cost += c1

    pi /= ite
    cost /= ite

    # Estimating expected utility
    util = 0
    for l in range(n+1):
        util += comb(n, l) * pi**l * (1 - pi)**(n - l) * (l * p1 - cost)

    return util
```

```python
def compute_estimated_utility(t1, p1, lambda1, c11, c21, c31, n, T):
    def lambda123(t):
        return 0.256 * t ** 0.837
    # Generating random variables for j=1
    e1 = lambda123(t1)
    eT = lambda123(T)
    q1 = eT - e1
    
    # Generating random variables for j=2,3
    ts = np.array([1000, 1000])
    ps = np.array([9000, 9000])
    qs = lambda123(T) - lambda123(ts)

    # Generating buyer random variables
    w = np.array([1/3, 1/3, 1/3])
    rho = 1

    # Computing utility
    u1 = 1 - np.exp(-rho * (-w[0] * t1 / T - w[1] * p1 / 15000 - w[2] * q1 / eT))
    u2 = 1 - np.exp(-rho * (-w[0] * ts[0] / T - w[1] * ps[0] / 15000 - w[2] * qs[0] / eT)) - u1
    u3 = 1 - np.exp(-rho * (-w[0] * ts[1] / T - w[1] * ps[1] / 15000 - w[2] * qs[1] / eT)) - u1

    # Computing probability of choice
    pi = 1 / (1 + np.exp(u2) + np.exp(u3))

    # Estimating the cost
    cost = c11 * t1 + c21 * e1 + c31 * (eT - e1)

    # Estimating expected utility
    util = 0
    for l in range(n+1):
        util += comb(n, l) * pi**l * (1 - pi)**(n - l) * (l * p1 - cost)

    return util
```

```python
# Example rate function lambda1(t)
def lambda1(t):
    return 0.256 * t ** 0.837
    
c11 = 0.5
c21 = 1
c31 = 5
n = 10  
T = 2000

# Example usage
t1 = 100
p1 = 4000
util = compute_expected_utility(t1, p1, lambda1, c11, c21, c31, n, T)
print("Expected Utility:", util)
util = compute_estimated_utility(t1, p1, lambda1, c11, c21, c31, n, T)
print("Estimated Utility:", util)
```

```python
import sympy as sp

def symbol_estimated_utility(t1, p1, c11, c21, c31, n, T):
    e1 = 0.256 * t1 ** 0.837
    eT = 0.256 * T ** 0.837
    q1 = eT - e1

    ts = [1000, 1000]
    ps = [9000, 9000]
    qs = [0.256 * T ** 0.837- 0.256 * tsi ** 0.837 for tsi in ts]


    w = [1/3, 1/3, 1/3]
    rho = 1

    u1 = 1 - sp.exp(-rho * (-w[0] * t1 / T - w[1] * p1 / 15000 - w[2] * q1 / eT))
    u2 = 1 - sp.exp(-rho * (-w[0] * ts[0] / T - w[1] * ps[0] / 15000 - w[2] * qs[0] / eT)) - u1
    u3 = 1 - sp.exp(-rho * (-w[0] * ts[1] / T - w[1] * ps[1] / 15000 - w[2] * qs[1] / eT)) - u1

    pi = 1 / (1 + sp.exp(u2) + sp.exp(u3))

    cost = c11 * t1 + c21 * e1 + c31 * (eT - e1)

    util = 0
    for l in range(n+1):
        util += comb(n, l) * pi**l * (1 - pi)**(n - l) * (l * p1 - cost)

    return util
```

```python
# Define symbolic variables
t1, p1 = sp.symbols('t1 p1')
n = 10
T_val = 2000
c11_val = 0.5
c21_val = 1
c31_val = 5

derivative = sp.diff(symbol_estimated_utility(t1, p1, c11_val, c21_val, c31_val, n, T_val), t1)
t1_solution = sp.solve(derivative, t1)
t1_solution
```

```python
# Define symbolic variables
t1, p1 = sp.symbols('t1 p1')
n = 10
T = 2000
c11 = 0.5
c21 = 1
c31 = 5

# Compute the derivative with respect to t1 and p1
derivative_t1 = sp.diff(symbol_estimated_utility(t1, p1, c11, c21, c31, n, T), t1)
derivative_p1 = sp.diff(symbol_estimated_utility(t1, p1, c11, c21, c31, n, T), p1)

# Solve the equations for critical points
critical_points = sp.solve((derivative_t1, derivative_p1), (t1, p1))

# Evaluate the function at each critical point to find the maximum
max_value = float('-inf')
max_point = None
for point in critical_points:
    value = symbol_estimated_utility(*point, c11_val, c21_val, c31_val, n_val, T_val)
    if value > max_value:
        max_value = value
        max_point = point

# Print the maximum value and corresponding point
print("Maximum value:", max_value)
print("Point of maximum (t1, p1):", max_point)
```
